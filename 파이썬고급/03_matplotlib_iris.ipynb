```python
features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
plt.figure(figsize=[12,8])

for i in range(4):
    plt.subplot(2,2,i+1)
    r1 = iris[:,i].min()-0.1
    r2 = iris[:,i].max()+0.1
    
    plt.hist(iris[:,i][:50], bins=np.linspace(r1,r2,21), alpha=0.2)
    plt.hist(iris[:,i][50:100], bins=np.linspace(r1,r2,21), alpha=0.2)
    plt.hist(iris[:,i][100:], bins=np.linspace(r1,r2,21), alpha=0.2)
    if i==0: plt.legend(['Setosa', 'Versicolor', 'Virginica'])
    plt.title(features[i])
```

### 산점도
- 두 속성 간의 상관관계를 알아본다.

```python
y
```

- 색상표 : https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

```python
features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
col1 = 0
col2 = 2

plt.scatter(X[:,col1], X[:,col2], c=y, alpha=0.5, cmap='spring')
plt.xlabel(features[col1])
plt.ylabel(features[col2])
plt.title('Petal Length vs Petal Width')
plt.colorbar()
```

```python
features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

n = 1

plt.figure(figsize=[12,12])

for col1 in range(3):
    for col2 in range(col1+1,4):
        plt.subplot(3,2,n)
        plt.scatter(X[:,col1], X[:,col2], c=y, s=y*50+50, alpha=0.5, cmap='summer')
        plt.xlabel(features[col1])
        plt.ylabel(features[col2])
        n = n+1
```
